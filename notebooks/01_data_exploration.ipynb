```python
# Data handling
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
sns.set_style("whitegrid")
```

```python
data = pd.read_csv('../data/raw/train.csv')
data.head()
```

```python
print(f"Dataset shape: {data.shape}")
```

```python
data.info()
```

```python
data.count()
```

```python
print("Columns:\n")
print(data.columns.tolist())

print("Data Types:\n")
print(data.dtypes)

print("Null Counts:\n")
print(data.isnull().sum())
```

```python
# Toxic label columns
label_cols = data.columns[2:]  # First two columns are 'id' and 'comment_text'

print("Label Distribution:\n")

for col in label_cols:
    print(f"{col}")
    print(data[col].value_counts())
    print("-" * 40)

# Rows with NO toxic labels
data["no_label"] = (data[label_cols].sum(axis=1) == 0).astype(int)

clean_count = data["no_label"].sum()

print(f"\nTotal CLEAN comments (all labels = 0): {clean_count}")
print(f"Total TOXIC comments: {len(data) - clean_count}")
```

```python
# Count how many toxic labels each comment has
data["toxic_count"] = data[label_cols].sum(axis=1)

# Distribution
toxic_count_distribution = data["toxic_count"].value_counts().sort_index()

print("Toxic Label Combination Distribution:\n")
print(toxic_count_distribution)
```

```python
plt.figure(figsize=(8, 5))

sns.barplot(
    x=toxic_count_distribution.index,
    y=toxic_count_distribution.values
)

plt.title("Number of Toxic Labels per Comment")
plt.xlabel("Toxic Label Count")
plt.ylabel("Number of Comments")

plt.show()
```

```python
# Word count
data["text_length"] = data["comment_text"].apply(
    lambda x: len(str(x).split())
)

print("Text Length Statistics:\n")

print("Minimum:", data["text_length"].min())
print("Maximum:", data["text_length"].max())
print("Mean:", round(data["text_length"].mean(), 2))
print("Median:", data["text_length"].median())
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(data["text_length"], bins=50)

plt.title("Comment Word Count Distribution")
plt.xlabel("Word Count")
plt.ylabel("Frequency")

plt.show()
```

```python
# Clean comments
clean_samples = data[data["toxic_count"] == 0]["comment_text"].sample(3, random_state=42)

print("CLEAN COMMENT SAMPLES:\n")

for i, comment in enumerate(clean_samples, 1):
    print(f"{i}. {comment}\n")
    print("-" * 80)
```

```python
# Toxic comments
toxic_samples = data[data["toxic_count"] > 0]["comment_text"].sample(3, random_state=42)

print("TOXIC COMMENT SAMPLES:\n")

for i, comment in enumerate(toxic_samples, 1):
    print(f"{i}. {comment}\n")
    print("-" * 80)
```
